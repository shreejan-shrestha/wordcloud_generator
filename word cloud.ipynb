```python
import pathlib
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
path = pathlib.Path.cwd() / 'interstellar.txt' # add a text file to the path
```

```python
text = path.read_text()
```

```python
wordcloud = WordCloud(background_color = "white").generate(text)
plt.figure(figsize = [10,10])
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.savefig("wordcloud_1.png", facecolor = "w", format = "png")
plt.show()
```

```python
mask = np.array(Image.open('img_1.jpg'))
stopwords = set(STOPWORDS)
plt.imshow(mask)
```

```python
wordcloud_1 = WordCloud(stopwords = stopwords, background_color = "white", mode = "RGBA").generate(text)
image_colors = ImageColorGenerator(mask)
plt.figure(figsize = [10,10])
plt.imshow(wordcloud_1.recolor(color_func=image_colors), interpolation = 'bilinear')
plt.axis("off")
plt.savefig('wordcloud_2.png', dpi = 1200, facecolor = "none", format = "png")
plt.show()
```
